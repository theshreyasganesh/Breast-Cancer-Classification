```python
import os 
from tqdm import tqdm 
```

```python
dir = '/home/mithilesh/jupyter/MINI-DDSM-Complete-JPEG-8/Cancer/'
```

```python
original_images = []
mask_images = []
original_images_f_name = []
mask_images_f_name = []
for root, directory, file in tqdm(os.walk(dir)):
    for f in file:
        if f.endswith('.jpg'):
            if 'MASK2' in f:
                continue
            if 'Mask' in f:
                mask_images_f_name.append(f)
                mask_images.append(root + '/' + f)
            else:
                original_images_f_name.append(f)
                original_images.append(root + '/' + f)
```

```python
new_original_images=[]
```

```python
og=original_images[0].split('/')
```

```python
len(og)
```

```python
for original in original_images_f_name:
    og=original.split('.')
    og=og[0]+'.'+og[1]
    for image in mask_images_f_name:
        if(og in image):
            new_original_images.append(original)
```

```python
final_original_images=[]
final_mask_images=[]
```

```python
for image in new_original_images:
    for imag_dir in original_images:
        if image in imag_dir:
            final_original_images.append(imag_dir)
```
