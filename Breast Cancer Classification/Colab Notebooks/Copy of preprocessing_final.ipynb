```python
def read_img(filename):
    image = io.imread(filename)
    return image
```

```python
def original_gray(original):  #function to convert images from bgr to grayscale
    gray_img = cv2.cvtColor(original,cv2.COLOR_BGR2GRAY)
    return gray_img
```

```python
def median_blur(denoise):
    denoise = cv2.medianBlur(denoise,3)
    return denoise
```

```python
def clahe(median_img):
    clahe = cv2.createCLAHE(clipLimit = 5)
    clahe_img = clahe.apply(median_img) + 0
    return clahe_img
```

```python
def preprocess(filename):
  img = read_img(filename)
  gray = original_gray(img)
  median = median_blur(gray)
  final = clahe(median)
  return final
```

```python
images_directory = '/content/drive/MyDrive/ultrasound/images_u/normal_cases'
pec_removed = '/content/drive/MyDrive/ultrasound/processed_us/normal_cases'
for root, directory, file in os.walk(images_directory):
  for f in file:
    
      pec_removed_img = preprocess(root + '/' + f)
      preprocess(root + '/' + f)
      pec_removed_img_path = pec_removed + '/' + f
      cv2.imwrite(pec_removed_img_path,255*pec_removed_img)
      #imsave(pec_removed_img_path,pec_removed_img)
      print(pec_removed_img_path)
   
```
