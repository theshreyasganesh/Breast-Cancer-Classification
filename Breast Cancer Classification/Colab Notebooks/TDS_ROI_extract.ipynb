```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("/content/drive/MyDrive/dump/mdb001.png")
plt.imshow(image)
plt.show()
```

```python
hsvImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsvImage,cmap='gray')
plt.show()
```

```python
lower_blue = np.array([100,150,0],np.uint8)
upper_blue = np.array([140,255,255],np.uint8)
mask = cv2.inRange(hsvImage, lower_blue, upper_blue)
```

```python
plt.imshow(mask)
plt.show()
```

```python
image = cv2.imread("/content/drive/MyDrive/dump/mdb001.png")
scam = cv2.imread("/content/drive/MyDrive/dump/mdb001.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3,3), np.uint8)
img_erosion = cv2.erode(image, kernel, iterations=2)
img_dilation = cv2.dilate(img_erosion, kernel, iterations=1)
#plt.imshow(img_erosion)
plt.imshow(img_dilation,cmap='gray')
plt.show()
```
