```python
import os 
from tqdm import tqdm 
```

```python
dir = '/content/drive/My Drive/mini_DDSM/Benign'
```

```python
original_images = []
mask_images = []
original_images_f_name = []
mask_images_f_name = []
for root, directory, file in tqdm(os.walk(dir)):
  for f in file:
    if f.endswith('.jpg'):
      if 'MASK2' in f:
        continue
      if 'Mask' in f:
        mask_images_f_name.append(f)
        mask_images.append(root + '/' + f)
      else:
        original_images_f_name.append(f)
        original_images.append(root + '/' + f)
```

```python
original = []
mask = []
for i in original_images_f_name:
  temp_1 = i.split('.')[0]
  temp_2 = i.split('.')[1]
  li = [temp_1,temp_2]
  temp = ('.').join(li)
  temp_mask = temp+'_Mask'
  mask.append(temp_mask)
  print(i,temp,temp_mask)
```

```python
images_with_masks = []
for i in mask:
  #print(i)
  for img in mask_images:
    if i in img:
      images_with_masks.append(i)
print(len(images_with_masks))
```

```python
amaze_images = []
for i in images_with_masks:
  temp = i.replace('_Mask','')
  amaze_images.append(temp)

proper_images = []
for i in amaze_images:
  for img in original_images:
    if i in img:
      proper_images.append(img)
print(len(proper_images))
```
